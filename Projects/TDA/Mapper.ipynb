```python
import yfinance as yf
import kmapper as km
import pandas as pd
from kmapper.jupyter import display
import umap
import sklearn
import sklearn.manifold as manifold
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/samuelsalander/Downloads/s-and-p-500-companies-main/data/constituents.csv')
#File with list of tickers on S and P 500, will be attached in the project folder, download and change filename
# as needed
tickers=df['Symbol'].tolist()
startt="2020-01-01" #Input desired start and ending dates, if end is in the future or start is beyond yfinance's
#library, an error will be thrown
endt="2022-04-02"
first=yf.download(tickers, start=startt, end=endt)
data_close = first['Adj Close'].dropna(axis='columns') #Get the data for those tickers and times and look at adjusted 
#closing prices
data = data_close.to_numpy()
data= data-np.mean(data, axis=0)/np.std(data,axis=0) #Normalize data
data=data.transpose()
stock_return=(data_close.to_numpy().transpose()[:,498]-data_close.to_numpy().transpose()[:,0])/data_close.to_numpy().transpose()[:,0]
#Calculate percent return of stock over the selected period
mapper = km.KeplerMapper(verbose=1) #initialize mapper
projected_data= mapper.fit_transform(data, projection=[manifold.Isomap(n_components=100, n_jobs=1), umap.UMAP(n_components=2, random_state=1)])
#Project the data with the mapper and UMAP
G=mapper.map(projected_data, data, clusterer=sklearn.cluster.DBSCAN(metric="cosine")) #Use default clusterer
#on sklearn to perform the algorithm
mapper.visualize(G,
                 path_html='/Users/samuelsalander/Downloads/mapper.html', #Change for your needs
                 title='s_and_p_mapper',
                custom_tooltips=data_close.columns.to_numpy(),
                color_values=np.log(stock_return+1), #Colored by either average, std, sum, max, or min log returns
                 #of data in each node
                color_function_name= 'Log Percent Returns',
                node_color_function=np.array(['average', 'std', 'sum', 'max', 'min']))

#You should now see a file with the title and filepath you specified above. Clicking on it should open a brower
#window to look at the result

```
